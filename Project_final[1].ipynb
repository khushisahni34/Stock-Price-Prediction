```python
#Feature Engineering

#Feature selection
data =df1[['Date','Open','High','Low','Volume','VWAP']]

#Setting index column as Date
df1.set_index("Date", drop=False, inplace=True)
#This means that date column cannot be dropped and all changes will be made to original data
```

```python
#Generating new features -Lag and Window features
df1.reset_index(drop=True, inplace=True)
lag_features =["Open","High","Low","VWAP"]
window1= 3 
window2= 7
 #Created a rolling window of 3days
df1_rolled_3d =df1[lag_features].rolling(window=window1, min_periods=0) #min_periods =0 specify that all observations will e included
df1_mean_3d= df1_rolled_3d.mean().shift(1).reset_index().astype(np.float32) 
#Rolling Mean (Moving Average):average of the values within the window.

#Rolling Standard Deviation:Measures the dispersion or volatility of values within the window.
#It shows that a 7-day rolling standard deviation indicates how much the stock price varies from the 3-day moving average.

df1_rolled_7d =df1[lag_features].rolling(window=window2, min_periods=0) #Created a rolling window of 7 days
df1_std_7d =df1_rolled_7d.std().shift(1).reset_index().astype(np.float32)

for feature in lag_features:
    df1[f"{feature}_mean_lag{window1}"] = df1_mean_3d[feature].astype('float32')
    df1[f"{feature}_std_lag{window2}"] = df1_std_7d[feature].astype('float32')
```

```python
#Setting all Null/NaN values of feature as mean of that feature
df1.fillna(df1.select_dtypes(include=[np.number]).mean(), inplace=True)
```

```python
#Setting Date as index and viewing the Dataset
df1.set_index("Date",drop=False,inplace=True)
df1.head()
```

```python
#EXtracting date time features from the dataset
df1.Date =pd.to_datetime(df1.Date, format="%Y-%m-%d")
df1["month"] =df1.Date.dt.month
df1["day"] =df1.Date.dt.day
df1["day_of_week"] =df1.Date.dt.dayofweek
#dayofweek assigns integers 1,2,.... to corresponding days monday,tuesday....
df1.head()
```

```python
#Removing Unneccesary features
df1= df1.drop(columns=["Symbol"])
df1= df1.drop(columns=["Series"])
print(df1.columns)
```

```python
df1.shape
```

```python
#Model building 
#Importing all the necessary libraries 

from sklearn.model_selection import train_test_split
!pip install xgboost
import xgboost as xgb
from xgboost import XGBRegressor
from xgboost import plot_importance

import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Define features (X) and target variable (y)
X = df1[['Open_mean_lag3', 'Open_std_lag7', 'High_mean_lag3', 'High_std_lag7',
       'Low_mean_lag3', 'Low_std_lag7', 'VWAP_mean_lag3', 'VWAP_std_lag7', 'month', 'day', 'day_of_week']]
y = df1['Close']

```

```python
# Splitting the data into training and testing sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
#Test_size =0.2 means 80% data is training and 20% is testing data
```

```python
# Define the XGBoost model
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.01)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(14,7))

plt.plot(y_train.index, y_train, label='Actual Prices',color='m')

plt.plot(y_test.index, y_pred, label='Predicted Prices',color='b')

plt.xlabel('Date')
plt.ylabel('Nifty 50 Closing Price')
plt.title('Actual vs Predicted Stock Prices')
plt.legend()
plt.show()
```

```python
# Evaluating the model built using
#Mean absolute error(MAE), Mean Squared Error(MSE) and R-squared

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
# Calculate r2_score - indicates the accuracy of the model
r_squared = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
print(f"R-squared: {r_squared}")
```
